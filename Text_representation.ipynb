```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
import seaborn as sns
```

```python
text = ["computational graph is an abstraction that models mathematical expressions",
        "oneÂ­hot representation for a phrase, sentence, or a document is simply a logical OR"]

one_hot = CountVectorizer(binary=True)
one_hot_counter = one_hot.fit_transform(text).toarray()
sns.heatmap(one_hot_counter, annot=True, cbar=False, yticklabels= ['Sentence 1', 'Sentence 2'])
```

```python
vector = TfidfVectorizer()
counter_vector = vector.fit_transform(text).toarray()
sns.heatmap(counter_vector, annot= True,  xticklabels='vocab',yticklabels=['Sentence 1','Sentence 2'],  linewidths=2, )
```
